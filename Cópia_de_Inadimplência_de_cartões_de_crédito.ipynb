#0. Introdução


Temos uma base com informações de inadimplência de 30000 Clientes
de cartão de Crédito.


Objetivo:

- Analisar os tipos de públicos que temos na carteira.
- Se a saúde da carteira de clientes está melhorando ou piorando com o
tempo.
- Possíveis problemas na concessão do crédito e/ou cobrança.


O que é Crédito?


Crédito significa confiança ou segurança em alguma coisa.
Crédito comercial ou bancário é definido como ato de vontade de alguém
destacar ou ceder, temporariamente, parte de seu patrimônio a um
terceiro, com a expectativa de que esta parcela volte à sua posse
integralmente após tempo estipulado.

O que é Inadimplência?

Inadimplência se refere ao não cumprimento de obrigações financeiras,
ou seja, a pessoa não pagou as dívidas dentro do prazo acordado.
Principais causas de inadimplência:
- Desemprego
- Despesas inesperadas
- Falta de planejamento financeiro


Conceitos


• FPD (First Payment Default): Atraso do primeiro pagamento em um empréstimo ou financiamento.
• OVER: indicador de dias de atraso atual. EX: OVER15 = atrasos acima de 15 dias.
• EVER: indicador de atraso máximo do contrato. EX: EVER30 = se teve algum atraso acima de 30 dias durante o
contrato.
• MOB (Months on Book): Quantidade de meses desde a concessão de crédito.
• SAFRA: Mês da concessão de crédito


Risco de Crédito


• Possibilidade de perdas financeiras:
• Não cumprimento das obrigações;
• Desvalorização dos ativos;
• Redução de remunerações e de ganhos esperados.


Provisionamento



PDD (Provisão para devedores duvidosos):


• Reserva de capital que a organização faz para lidar com os
imprevistos de um possível não pagamento. Ou seja, separar
dinheiro para não ter problemas ocasionados pela inadimplência.

### a)Análise SWOT


Strengths (Forças)
Base de dados robusta: 30.000 clientes fornecem uma amostra significativa para análise.

Indicadores claros: Métricas como FPD, OVER, EVER e MOB permitem uma avaliação detalhada.

Possibilidade de segmentação: Dados como SAFRA e MOB permitem analisar tendências temporais.

Gestão de risco estruturada: Conceitos como PDD mostram uma preocupação com provisionamento.

Weaknesses (Fraquezas)
Possível viés na amostra: Dados históricos podem não refletir mudanças recentes no mercado.

Falta de variáveis contextuais: Fatores como desemprego ou crises econômicas podem impactar inadimplência.

Dependência de métricas tradicionais: OVER e EVER podem não capturar padrões comportamentais mais sutis.

Opportunities (Oportunidades)
Modelos preditivos: Usar machine learning para prever inadimplência com base em histórico.

Segmentação avançada: Identificar clusters de clientes (ex: bons pagadores repentinamente inadimplentes).

Ajuste na política de crédito: Reduzir concessões para perfis de alto risco com base em dados.

Melhoria na cobrança: Estratégias personalizadas conforme perfil de inadimplência.

Threats (Ameaças)
Mudanças macroeconômicas: Crises podem aumentar inadimplência além do previsto.

Concorrência: Bancos e fintechs podem atrair bons pagadores, deixando a carteira mais arriscada.

Regulação: Novas leis podem limitar critérios de concessão ou cobrança.

```python
# Dados fornecidos (simulando df["Over"].value_counts())
contagem = df["Over"].value_counts().sort_index()  # Ordena por Over30, Over60, Over90, Nenhum
porcentagens = (contagem / contagem.sum()) * 100  # Calcula as %

# Gráfico de barras com seaborn
plt.figure(figsize=(10, 6))
ax = sns.barplot(x=contagem.index, y=contagem.values, palette="Set2")

# Adiciona título e labels
plt.title("Distribuição de Atrasos na Carteira de Crédito", fontsize=14, pad=20)
plt.ylabel("Número de Clientes", fontsize=12)
plt.xlabel("Dias em Atraso", fontsize=12)

# Adiciona as porcentagens em cima das barras
for p, porcent in zip(ax.patches, porcentagens):
    ax.annotate(
        f"{porcent:.1f}%",
        (p.get_x() + p.get_width() / 2., p.get_height()),
        ha="center",
        va="center",
        xytext=(0, 10),
        textcoords="offset points",
        fontsize=10
    )

plt.show()
plt.show()
```

Principais Insights:


1. Concentração de Inadimplência nos Atrasos de 60 Dias (Over60)
Over60 (9.118 clientes) é a categoria com maior volume de inadimplentes, superando até mesmo os atrasos menores (Over30) e os mais graves (Over90).



Possíveis causas:


Pode indicar que muitos clientes enfrentam dificuldades financeiras após 60 dias, mas conseguem regularizar antes de chegar a 90 dias.

Pode refletir um problema específico na política de cobrança (ex.: falta de ações eficazes antes dos 60 dias).



2. Baixo Número de Atrasos Graves (Over90) em Comparação com Over60
Over90 (1.856 clientes) é menor que Over60 (9.118), o que sugere que:

Parte dos clientes em Over60 consegue quitar a dívida antes de virar Over90.

Pode haver uma recuperação significativa entre 60 e 90 dias (ex.: negociações, cobrança mais efetiva).

Ou, ainda, que muitos clientes em Over60 não evoluem para Over90 porque são cortados do crédito antes.

3. Atrasos Leves (Over30) São Menores que Over60 e Over90
Over30 (1.645) é a menor categoria, o que pode indicar:

Boa capacidade de pagamento inicial, pois poucos clientes atrasam 30 dias.

Política de cobrança eficiente nos primeiros 30 dias (ex.: lembretes automáticos, juros baixos no início).

Ou que muitos clientes pulam direto para Over60, sem ficar muito tempo em Over30.

4. Comparação entre Clientes Adimplentes e Inadimplentes
Clientes sem atraso (Nenhum) = 17.381

Clientes com atraso (Over30+60+90) = 12.619

Taxa de inadimplência = (12.619 / (17.381 + 12.619)) ≈ 42%

Isso é alto, indicando que quase 42% da carteira está com algum nível de atraso.

## 4.1 Information Value

```python

def calculate_iv(df, feature, target, woe_tables):
    """
    Calcula o Information Value (IV) com tratamento robusto para evitar valores infinitos
    """
    try:
        # Remove linhas com valores nulos
        df_clean = df[[feature, target]].dropna()

        # Verifica se há dados suficientes
        if len(df_clean) < 2 or df_clean[target].nunique() < 2:
            return 0

        # Agrupa os dados
        grouped = df_clean.groupby(feature)[target].agg(['count', 'sum'])
        grouped.columns = ['Total', 'Inadimplentes']
        grouped['Não_Inadimplentes'] = grouped['Total'] - grouped['Inadimplentes']

        # Adiciona pequena constante para evitar divisão por zero (suavização de Laplace)
        eps = 1e-10
        total_inad = grouped['Inadimplentes'].sum() + eps
        total_ninad = grouped['Não_Inadimplentes'].sum() + eps

        # Cálculo seguro das porcentagens
        grouped['%Inad'] = (grouped['Inadimplentes'] + eps) / total_inad
        grouped['%NInad'] = (grouped['Não_Inadimplentes'] + eps) / total_ninad

        # Cálculo do WOE com tratamento especial
        with np.errstate(divide='ignore', invalid='ignore'):
            woe = np.log(grouped['%Inad'] / grouped['%NInad'])
            grouped['WOE'] = np.where(
                (grouped['Inadimplentes'] == 0) | (grouped['Não_Inadimplentes'] == 0),
                0,  # Atribui zero quando um dos lados for zero
                woe
            )

        # Cálculo do IV com limitação
        grouped['IV'] = (grouped['%Inad'] - grouped['%NInad']) * grouped['WOE']
        grouped['IV'] = np.clip(grouped['IV'], -10, 10)  # Limita valores extremos

        # Armazena resultados
        woe_tables[feature] = grouped.reset_index()

        iv_total = grouped['IV'].sum()
        return max(0, min(iv_total, 10))  # Retorna IV entre 0 e 10

    except Exception as e:
        print(f"Erro ao calcular IV para {feature}: {str(e)}")
        return 0

def classificar_iv(valor):
    """Classifica o IV de acordo com a escala padrão"""
    valor = min(max(valor, 0), 10)  # Garante que está entre 0 e 10
    if valor < 0.02: return 'Inútil'
    elif valor < 0.1: return 'Fraco'
    elif valor < 0.3: return 'Médio'
    elif valor < 0.5: return 'Forte'
    else: return 'Muito Forte'

# Exemplo de uso completo:
if __name__ == "__main__":
    # 1. Carregar os dados
    df = pd.read_excel('/content/base_inad_v1.xlsx')

    # 2. Pré-processamento da variável target
    df['Inadimplente'] = df['Over'].apply(lambda x: 1 if x in ['Over30', 'Over60', 'Over90'] else 0)
    target = 'Inadimplente'

    # 3. Lista de features para análise
    features = ['Limite', 'Idade', 'Fatura', 'Educação', 'Gênero', 'Estado_civil']

    # 4. Dicionário para armazenar as tabelas WOE
    woe_tables = {}
    iv_results = {}

    # 5. Calcular IV para todas as features
    for feature in features:
        iv_results[feature] = calculate_iv(df, feature, target, woe_tables)

    # 6. Criar DataFrame com resultados
    iv_df = pd.DataFrame.from_dict(iv_results, orient='index', columns=['IV'])
    iv_df = iv_df.sort_values('IV', ascending=False)
    iv_df['Classificação'] = iv_df['IV'].apply(classificar_iv)

    # 7. Visualização dos resultados
    print("="*60)
    print("Análise de Poder Preditivo (Information Value)")
    print("="*60)
    print(iv_df.to_string())

    # 8. Exemplo de visualização gráfica
    plt.figure(figsize=(10, 6))
    bars = plt.bar(iv_df.index, iv_df['IV'], color='steelblue')

    # Adiciona valores nas barras
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2., height,
                f'{height:.4f}',
                ha='center', va='bottom')

    # Linhas de referência
    for val, color, label in [(0.02, 'red', 'Inútil'),
                            (0.1, 'orange', 'Fraco'),
                            (0.3, 'green', 'Médio'),
                            (0.5, 'blue', 'Forte')]:
        plt.axhline(val, color=color, linestyle='--', linewidth=1)
        plt.text(len(iv_df)-0.3, val+0.02, label, color=color)

    plt.title('Information Value por Variável')
    plt.ylabel('IV')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
# Carregar o DataFrame
df = pd.read_excel('/content/base_inad_v1.xlsx')

# Criar colunas binárias para cada categoria de 'Over'
df['Over30'] = (df['Over'] == 'Over30').astype(int)
df['Over60'] = (df['Over'] == 'Over60').astype(int)
df['Over90'] = (df['Over'] == 'Over90').astype(int)

# Função para criar faixas e calcular métricas
def calcular_metricas(df, coluna, faixas, labels):
    # Verificar se as faixas estão em ordem crescente
    if not np.all(np.diff(faixas) > 0):
        raise ValueError("As faixas devem estar em ordem crescente.")

    # Usar numpy.digitize para criar índices das faixas
    indices = np.digitize(df[coluna], bins=faixas, right=True)

    # Mapear índices para labels e tratar valores fora das faixas
    df['Faixa'] = [labels[i-1] if i <= len(labels) else 'Outros' for i in indices]

    # Agrupar por faixa e calcular métricas para cada categoria de inadimplência
    grouped = df.groupby('Faixa').agg(
        Total=('Over30', 'size'),  # Contagem total de clientes
        Over30=('Over30', 'sum'),  # Contagem de Over30
        Over60=('Over60', 'sum'),  # Contagem de Over60
        Over90=('Over90', 'sum')   # Contagem de Over90
    ).reset_index()

    # Calcular % para cada categoria de inadimplência
    grouped['% Over30'] = (grouped['Over30'] / grouped['Total'] * 100).round(1)
    grouped['% Over60'] = (grouped['Over60'] / grouped['Total'] * 100).round(1)
    grouped['% Over90'] = (grouped['Over90'] / grouped['Total'] * 100).round(1)

    # Substituir NaN por 0
    grouped.fillna(0, inplace=True)

    return grouped


# Definir faixas para cada variável (CORRIGIDAS)
faixas_limite = [10000, 30000, 50000, 100000, float('inf')]  # Faixas ajustadas para Limite (ordem crescente)
labels_limite = ['10.000 - 30.000', '30.001 - 50.000', '50.001 - 100.000', '> 100.000']

faixas_fatura = [0, 500, 1500, 3000, float('inf')]  # Faixas ajustadas para Fatura
labels_fatura = ['0 - 500', '501 - 1.500', '1.501 - 3.000', '> 3.000']

faixas_idade = [18, 26, 38, 50, float('inf')]  # Faixas ajustadas para Idade
labels_idade = ['18 - 26', '27 - 38', '39 - 50', '> 50']

# Calcular métricas para cada variável
limite_df = calcular_metricas(df, 'Limite', faixas_limite, labels_limite)
fatura_df = calcular_metricas(df, 'Fatura', faixas_fatura, labels_fatura)
idade_df = calcular_metricas(df, 'Idade', faixas_idade, labels_idade)

# Consolidar resultados
results = {
    'Limite': limite_df,
    'Fatura': fatura_df,
    'Idade': idade_df
}

# Exibir tabelas por variável
print("="*80)
print("ANÁLISE COMPLETA DE INADIMPLÊNCIA POR VARIÁVEL")
print("="*80)
for var, df_var in results.items():
    print(f"\n{var}")
    print("-"*60)
    # Formatar os números de porcentagem com %
    df_var_formatted = df_var.copy()
    df_var_formatted['% Over30'] = df_var_formatted['% Over30'].apply(lambda x: f"{x}%")
    df_var_formatted['% Over60'] = df_var_formatted['% Over60'].apply(lambda x: f"{x}%")
    df_var_formatted['% Over90'] = df_var_formatted['% Over90'].apply(lambda x: f"{x}%")
    print(df_var_formatted[['Faixa', 'Total', 'Over30', '% Over30',
                           'Over60', '% Over60', 'Over90', '% Over90']].to_string(index=False))


# Visualização gráfica
plt.figure(figsize=(15, 10))

# Cores para cada categoria
colors = ['#FFB3BA', '#FFDFBA', '#B3E6FF']  # Rosa, Laranja, Azul

# Lista de categorias e seus percentuais
categories = ['% Over30', '% Over60', '% Over90']

for i, (var, df_var) in enumerate(results.items(), 1):
    plt.subplot(2, 2, i)

    # Plotar cada categoria separadamente
    bottom = 0
    for j, category in enumerate(categories):
        values = df_var[category].values
        plt.bar(df_var['Faixa'], values, color=colors[j], bottom=bottom, label=category)
        bottom += values

    # Adicionar legendas dentro do gráfico
    plt.legend(loc='upper left', bbox_to_anchor=(1, 1))

    # Adicionar valores em cada segmento da barra
    for idx, row in df_var.iterrows():
        x = idx
        y = 0
        for j, category in enumerate(categories):
            value = row[category]
            plt.text(x, y + value/2, f'{value}%', ha='center', va='center')
            y += value

    plt.title(f'Distribuição de Atraso por Faixa - {var}')
    plt.ylabel('% Clientes')
    plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
